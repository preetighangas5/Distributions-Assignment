**Question 5:** What is Kurtosis and its type?

**Answer: **Kurtosis is a statistical measure that describes the shape of a distribution's tails and peakedness relative to a normal distribution. It quantifies how much data is in the tails versus the center. Types include:

*   Mesokurtic: Kurtosis equals 3 (or excess kurtosis of 0),similar to
                normal distribution with moderate tails.
*  Leptokurtic: Kurtosis greater than 3 (excess > 0), with a sharper
                peak and heavier tails (more outliers).
   
*   Platykurtic: Kurtosis less than 3 (excess < 0), with a flatter
                 peak and lighter tails(fewer outliers).


**Question 6:** Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

**Answer: **The uniform distribution is a good model for rolling a fair die because each of the six outcomes (1 through 6) is equally likely, with a probability of 1/6, and there is no bias toward any particular result. This matches the definition of a discrete uniform distribution, where all possible values have the same probability density.

**Question 8 :**How does log transformation help in stabilizing variance and making data more normally distributed?

**Answer: **Log transformation helps stabilize variance by compressing large values more than small ones, which is useful for data where variance increases with the mean (e.g., in multiplicative or heteroscedastic processes). It also reduces positive skewness, making the data more symmetric and closer to a normal distribution, which improves the applicability of parametric statistical tests that assume normality.
